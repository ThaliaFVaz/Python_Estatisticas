```python
import random
import pandas as pd
import numpy as np
import plotly.express as px
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
#cria uma lista contendo todas as datas entre 01/01/2021 até 31/12/2021

dates = pd.date_range("2021/01/01", "2021/12/31", 365)
print(dates)
```

```python
#cria uma lista contendo uma série quase senoidal contendo também números nulos

def RandomOrNan(i):
    if random.randrange(0, 7) != 0:
        valor = (1+math.sin(i/5))*10 + random.uniform(0.0, 3.0)*(i/100)
        return round(valor, 2)
    else:
        return np.nan

randomsign = [RandomOrNan(i) for i in range (len(dates))]
print(randomsign)
```

```python
#Cria um arquivo no formato pandas dataframe contendo 2 colunas: as datas e os números criados acima

df = pd.DataFrame()
for i in range(len(dates)):
    serie = {'data':dates[i], 'valor':randomsign[i]}
    df = df.append(serie, ignore_index=True)

#mostra as primeiras 20 linhas do dataframe criado
df.head(20)
```

```python
#plota o dataframe

#fig = px.line(df, x = df.data, y = ['valor'], title="meuplot")

df.plot(x ='data', y=['valor'], figsize=(10, 5), kind = 'line', title="sinal original")
plt.show()
```

```python
df.to_csv(r'quase senoidal.csv', index = False)
```
